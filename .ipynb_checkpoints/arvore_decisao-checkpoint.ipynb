### Separando conjunto de dados de features e target

```python
X_dados = dados_diabetes.drop('diabetes', axis=1).values
X_dados
```

```python
y_dados = dados_diabetes['diabetes'].values
y_dados
```

### Criando função para balancear os dados

```python
# UnderSampler
def balanceamento_dados(X_dados, y_dados):
    under_sample = RandomUnderSampler(random_state=42)
    X_under, y_under = under_sample.fit_resample(X_dados, y_dados)

    tl = TomekLinks(sampling_strategy='all') # Ajuda a melhorar o balanceamento de dados
    X_under, y_under = tl.fit_resample(X_under, y_under)

    return X_under, y_under
```

### Aplicando técnmica de balanceamento na base de dados

```python
X_dados_balanceados, y_dados_balanceados = balanceamento_dados(X_dados, y_dados)
```

### Verificando balanceamento realizado

```python
print(len(X_dados_balanceados), len(y_dados_balanceados))
```

```python
# Verificar lista com dados balanceados
import collections
print(collections.Counter(y_dados_balanceados))
```

### Criando função para realizar separação entre o conjunto de treinamento 

```python
from sklearn.model_selection import train_test_split
```

```python
def separa_treino_teste(X_dados_balanceados, y_dados_balanceados):
    X_train, X_test, y_train, y_test = train_test_split(X_dados_balanceados, 
                                                        y_dados_balanceados,
                                                        random_state=42,
                                                        test_size=0.2) # 80% para teste e 20% para teste
    return X_train, X_test, y_train, y_test
```

```python
# Seprando conjuntos de treino e teste
X_train, X_test, y_train, y_test = separa_treino_teste(X_dados_balanceados, y_dados_balanceados)
```

```python
print(len(X_train), len(y_train))
```

```python
print(len(X_test), len(y_test))
```

### Criando árvore de decisão

```python
from sklearn import tree
algoritmo_arvore = tree.DecisionTreeClassifier()
modelo = algoritmo_arvore.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

### Criando lista com os nomes das features e dos rótulos

```python
nomes_features = ['gravidez', 'glicose', 'pressao_sangue', 'espessura_pele', 'insulina',
                      'IMC', 'predisposicao_diabetes', 'idade']

nomes_classes = ['NÃO DIABETICO', 'DIABETICO']
```

### Criando função para visualização da árvore gerada

```python
import graphviz
def visualiza_arvore(modelo, nomes_features, nomes_classes):
    arvore = tree.export_graphviz(modelo,
                                  feature_names=nomes_features,
                                  class_names=nomes_classes,
                                  rounded=True,
                                  special_characters=True,
                                  filled=True)
    graph = graphviz.Source(arvore)
    return graph
```

```python
# Visualiza a árvore completa
visualiza_arvore(modelo, nomes_features, nomes_classes)
```

### Calculando a acurária do modelo

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

### Criando função que gera features de importância para o modelo

```python
def gera_features_importantes(modelo):
    df = pd.DataFrame(modelo.feature_importances_, columns=['importancia'], 
                      index = ['gravidez', 'glicose', 'pressao_sangue', 'espessura_pele', 'insulina',
                               'IMC', 'predisposicao_diabetes', 'idade']).sort_values('importancia', ascending=False)

    return df
```

```python
print(gera_features_importantes(modelo))
```

### Criando função que gera árvore de decisão

```python
PATH_IMG = r'./dataset/'
```

```python
def cria_modelo(X_train, y_train, y_test, tamanho_arvore, nomes_features, nomes_classes, PATH_IMG):

    algoritmo_arvore = tree.DecisionTreeClassifier(max_depth=tamanho_arvore, random_state=42)
    modelo = algoritmo_arvore.fit(X_train, y_train)
    y_predito = modelo.predict(X_test)
    arvore = visualiza_arvore(modelo, nomes_features, nomes_classes)

    # Salva arquivo PDF
    arvore.render(f'{PATH_IMG}/modelo_tamanho_n{tamanho_arvore}')

    print(f''' Features com maior importância para o modelo:
            {gera_features_importantes(modelo)}
          ''')

    print('O valor da acurácia é: ', accuracy_score(y_test, y_predito))

    return arvore
```

### Gerando resultados com poda de árvore

```python
tamanho_arvore = 20
arvore = cria_modelo(X_train, y_train, y_test, tamanho_arvore, nomes_features, nomes_classes, PATH_IMG)
arvore
```
